```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

def read_plot_one_basin(b):
    inpath = '/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/Calib_HH_MOASMO_bigrange'

    filemet = f'{inpath}/level1_{b}_MOASMOcalib/ctsm_outputs/iter0_all_metric.csv'
    df_met0 = pd.read_csv(filemet)
    
    filemet = f'{inpath}/level1_{b}_MOASMOcalib/ctsm_outputs/iter1_all_metric.csv'
    df_met1 = pd.read_csv(filemet)
    
    fileint = f'{inpath}/level1_{b}_MOASMOcalib/param_sets/intermediate_output_iter1.pkl.npz'
    loaded_data = np.load(fileint)
    x_resample = loaded_data['x_resample']
    y_resample = loaded_data['y_resample']
    x = loaded_data['x']
    y = loaded_data['y']
    
    fig = plt.figure(figsize=[4, 4])
    
    plt.scatter(y[:,0], y[:,1], s=10, facecolor='black', edgecolors=None, label='real model iter-0')
    plt.scatter(y_resample[:,0], y_resample[:,1], s=10, edgecolors=None, facecolor='blue', alpha=0.5, label='emulator iter-1')
    plt.scatter(df_met1.values[:,0], df_met1.values[:,1], s=10, edgecolors=None, facecolor='red', alpha=0.5, label='real model iter-1')
    plt.title(f'basin {b}')
    plt.xlabel('metric 1')
    plt.ylabel('metric 2')
    plt.legend()
    plt.tight_layout()
    plt.show()
```
