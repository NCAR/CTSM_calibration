```python
# For a given parameter list, get the default parameters (mean value) of it
# this can be used when using new parameters for SA or calib
```

```python
import os, sys, subprocess, pickle
import numpy as np
import pandas as pd
import xarray as xr

# import MO-ASMO functions
path_MOASMO = '../MOASMO_support/'
sys.path.append(path_MOASMO)
from MOASMO_parameters import *
```

```python
file_parameter_list = 'CTSM_CAMELS_SA_param_240202.csv'
path_CTSM_case='/glade/work/guoqiang/CTSM_CAMELS/SA_HH_allbasins/level1' # Find a case which uses default parameters
```

```python
df_calibparam = read_parameter_csv(file_parameter_list)
param_names = df_calibparam['Parameter'].values
param_sources = df_calibparam['Source'].values
        
dfi = df_calibparam.copy()
param0 = []
for j in range(len(param_names)):
    pj = get_parameter_value_from_CTSM_case(param_names[j], param_sources[j], path_CTSM_case)
    pj = np.nanmean(pj)
    param0.append(pj)

dfi['Value'] = param0
```

```python
dfi.to_csv('addDefaultValue_'+file_parameter_list, index=False)
```
