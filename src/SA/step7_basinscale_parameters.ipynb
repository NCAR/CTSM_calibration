```python
# for each basin, the default parameter and parameter range are from PPE (global default)
# this script extract the true mean default parameters from the CTSM case (e.g., surface dataset, name list, param nc, etc)

import xarray as xr
import numpy as np
import pandas as pd
import glob, os, sys

sys.path.append('../MOASMO_support')
from MOASMO_parameters import read_save_load_all_default_parameters
```

# load data

```python
infile_basin_info = f'/glade/work/guoqiang/CTSM_CAMELS/data_mesh_surf/HillslopeHydrology/CAMELS_level1_basin_info.csv'
df_info = pd.read_csv(infile_basin_info)
```

```python
for basin_num in range(len(df_info)):

    idi = df_info.iloc[basin_num]['hru_id']

    paramfile = f'/glade/work/guoqiang/CTSM_CAMELS/data_paramcailb/ParamCalib_smallrange_{idi}.csv'
    paramfile_new = f'/glade/work/guoqiang/CTSM_CAMELS/data_paramcailb/ParamCalib_smallrange_basinparam_{idi}.csv'
    
    if os.path.isfile(paramfile_new):
        continue
    
    
    # load raw params
    df0 = pd.read_csv(paramfile)

    df0 = df0.rename(columns={'Default':'Default_glob', 'Lower':'Lower_glob', 'Upper':'Upper_glob'})

    # read default param
    path_CTSM_case = f'/glade/work/guoqiang/CTSM_CAMELS/Calib_HH_MOASMO/level1_{basin_num}'
    dfnew = read_save_load_all_default_parameters(paramfile, '', path_CTSM_case=path_CTSM_case, savefile=False)

    defa = np.array([np.mean(i) for i in dfnew['Value']])
    df0['Default'] = defa
    df0['Lower'] = (defa/df0['Default_glob'].values) * df0['Lower_glob'].values
    df0['Upper'] = (defa/df0['Default_glob'].values) * df0['Upper_glob'].values
    
    df0.to_csv(paramfile_new)
```
