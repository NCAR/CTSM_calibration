```python
# for MOSART grids, find the one corresponding to outlet
import pandas as pd
import os, sys, glob
import matplotlib.pyplot as plt
import geopandas as gpd
from shapely.geometry import Point, Polygon
import numpy as np
import xarray as xr
```

```python
file = '/glade/campaign/cesm/cesmdata/inputdata/rof/mosart/MOSART_routing_0.125nldas2_cdf5_c200727.nc'
ds_mo = xr.load_dataset(file)
```

```python
lon = ds_mo.longxy.values[0,:]
lat = ds_mo.latixy.values[:,0]
areaTotal = ds_mo.areaTotal.values
ID = ds_mo.ID.values  # Flatten in case of 2D array
dnID = ds_mo.dnID.values
fdir = ds_mo.fdir.values
frac = ds_mo.frac.values
rslp = ds_mo.rslp.values
rlen = ds_mo.rlen.values
rdep = ds_mo.rdep.values
rwid = ds_mo.rwid.values
rwid0 = ds_mo.rwid0.values
gxr = ds_mo.gxr.values
hslp = ds_mo.hslp.values
twid = ds_mo.twid.values
tslp = ds_mo.tslp.values
area = ds_mo.area.values
areaTotal2 = ds_mo.areaTotal2.values
nr = ds_mo.nr.values
nt = ds_mo.nt.values
nh = ds_mo.nh.values

polygons = []
data_dict = {
    'ID': [],
    'dnID': [],
    'fdir': [],
    'frac': [],
    'rslp': [],
    'rlen': [],
    'rdep': [],
    'rwid': [],
    'rwid0': [],
    'gxr': [],
    'hslp': [],
    'twid': [],
    'tslp': [],
    'area': [],
    'areaTotal': [],
    'areaTotal2': [],
    'nr': [],
    'nt': [],
    'nh': [],
}

res = 0.125

# Create polygons for each grid cell
for i in range(len(lat)):
    for j in range(len(lon)):
        if True:
            # Define corners of the polygon based on center points
            if j < len(lon) - 1 and i < len(lat) - 1:  # Ensure we do not go out of bounds
                corners = [
                    (lon[j] - res/2, lat[i] - res/2),     # Bottom-left
                    (lon[j + 1] - res/2, lat[i] - res/2), # Bottom-right
                    (lon[j + 1] - res/2, lat[i + 1] - res/2), # Top-right
                    (lon[j] - res/2, lat[i + 1] - res/2), # Top-left
                ]
                polygons.append(Polygon(corners))
    
                # Append corresponding data
                data_dict['ID'].append(ID[i, j])
                data_dict['dnID'].append(dnID[i, j])
                data_dict['fdir'].append(fdir[i, j])
                data_dict['frac'].append(frac[i, j])
                data_dict['rslp'].append(rslp[i, j])
                data_dict['rlen'].append(rlen[i, j])
                data_dict['rdep'].append(rdep[i, j])
                data_dict['rwid'].append(rwid[i, j])
                data_dict['rwid0'].append(rwid0[i, j])
                data_dict['gxr'].append(gxr[i, j])
                data_dict['hslp'].append(hslp[i, j])
                data_dict['twid'].append(twid[i, j])
                data_dict['tslp'].append(tslp[i, j])
                data_dict['area'].append(area[i, j])
                data_dict['areaTotal'].append(areaTotal[i, j])
                data_dict['areaTotal2'].append(areaTotal2[i, j])
                data_dict['nr'].append(nr[i, j])
                data_dict['nt'].append(nt[i, j])
                data_dict['nh'].append(nh[i, j])


# Create a GeoDataFrame
gdf = gpd.GeoDataFrame(data_dict, geometry=polygons)

# Define the coordinate reference system (CRS)
gdf.crs = "EPSG:4326"  # WGS 84

# Save to a shapefile
gdf.to_file("MOSART_routing_0.125nldas2_cdf5_c200727.gpkg")
```
