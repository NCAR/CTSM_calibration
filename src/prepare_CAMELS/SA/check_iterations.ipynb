```python
files = glob.glob('/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/SA_HH_allbasins/level1/ctsm_outputs/iter0_trial0/lnd/hist/*h1*.nc')
files.sort()
ds1 = xr.open_mfdataset(files)
```

```python
files = glob.glob('/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/SA_HH_allbasins/level1/ctsm_outputs/iter0_trial1/lnd/hist/*h1*.nc')
files.sort()
ds2 = xr.open_mfdataset(files)
```

```python
files = glob.glob('/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/SA_HH_allbasins/level1/ctsm_outputs/iter0_trial2/lnd/hist/*h1*.nc')
files.sort()
ds3 = xr.open_mfdataset(files)
```

```python
ds1.QRUNOFF.sel(time=slice('2004', '2009')).mean(dim='lndgrid').plot()
ds2.QRUNOFF.sel(time=slice('2004', '2009')).mean(dim='lndgrid').plot()
ds3.QRUNOFF.sel(time=slice('2004', '2009')).mean(dim='lndgrid').plot()
```
