```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import matplotlib
```

```python
def plot_one_basin(basin, figfile):

    path = f'/glade/campaign/cgd/tss/people/guoqiang/CTSMcases/CAMELS_Calib/Calib_all_HH_MOASMO/{basin}_MOASMOcalib/ctsm_outputs/'
    iternum = 9

    df = []
    for i in range(iternum):
        df.append(pd.read_csv(f'{path}/iter{i}_all_metric.csv'))


    # Choose a colormap from matplotlib.cm module (e.g., 'viridis', 'plasma', 'jet', etc.)
    colormap_name = 'cool'

    # Number of colors you want to obtain from the colormap
    num_colors = iternum

    # Create a numpy array of equally spaced values from 0 to 1 (representing the colormap range)
    values = np.linspace(0, 1, num_colors)

    # Get the colormap using the specified name
    colormap = plt.get_cmap(colormap_name)

    # Get the list of colors by evaluating the colormap at the specified values
    colors = [colormap(value) for value in values]
    
    
    font = {'size'   : 12}
    matplotlib.rc('font', **font)
    
    plt.figure(figsize=[7, 7])
    
    for i in range(iternum):
        plt.scatter(df[i]['metric1'], df[i]['metric2'], 20, label=f'iter{i}', color=colors[i])

    plt.xlabel("1 - KGE'")
    plt.ylabel("Max abs monthly error")
    plt.legend(loc='lower right')
    # plt.show()
    plt.savefig(figfile, dpi=600, facecolor='w', bbox_inches='tight',pad_inches = 0.1)
```

```python
plot_one_basin('level1_8', 'MOASMO_level1_8.png')
```
