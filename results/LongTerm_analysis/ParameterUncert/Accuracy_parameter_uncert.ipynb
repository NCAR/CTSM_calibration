# Parameters have similar performance in the calibration period (e.g., normKGE) can be different during the validation period and long term simulation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import os, sys, glob
```

# Load metrics

```python
# load defa metrics

csvs = ['evaluation_metrics_5y_test.csv', 'evaluation_metrics_Ally_test.csv',
        'evaluation_metrics_train.csv', 'evaluation_metrics_ally.csv'] # 5y test period, all year test (no train period), train, all years

df_defa = []
for basin in range(627):
    pathb = f'/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/LongTermSimu/Defa/level1_{basin}/'
    files = [f'{pathb}/{c}' for c in csvs]
    dfb = pd.concat(map(pd.read_csv, files))
    dfb['basin'] = basin
    dfb['eval'] = ['test_5y', 'test_ally', 'train', 'ally']
    dfb['param'] = 'Defa'
    df_defa.append(dfb)
df_defa = pd.concat(df_defa)
```

```python
# load optmz metrics
df_optmz = []
for basin in range(627):

    df_b = []
    # for r in range(1, 6):
    for r in [1, 3, 5, 10, 20]:
        pathb = f'/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/LongTermSimu/LSEallbasin/level1_{basin}/normKGEr{r}'
        files = [f'{pathb}/{c}' for c in csvs]
        if not os.path.isfile(files[0]):
            print('Warning! File does not exist for', basin)
            dfb = dfb.copy()
            dfb[:]=np.nan
            dfb['basin'] = basin
            dfb['eval'] = ['test_5y', 'test_ally', 'train', 'ally']
            dfb['param'] = f'normKGEr{r}'
        else:
            dfb = pd.concat(map(pd.read_csv, files))
            dfb['basin'] = basin
            dfb['eval'] = ['test_5y', 'test_ally', 'train', 'ally']
            dfb['param'] = f'normKGEr{r}'
        df_b.append(dfb)

    df_b = pd.concat(df_b)
    
    df_optmz.append(df_b)
df_optmz = pd.concat(df_optmz)
```

```python
# merge them into one df
df_all = pd.concat([df_defa, df_optmz])
```

# Comparison

```python
# train vs test
print('Defa')
for e in ['test_5y', 'test_ally', 'train', 'ally']:
    d1 = df_defa[df_defa['eval']==e]['kge'].values
    print(e, np.nanmean(d1), np.nanmedian(d1))


print('Optmz')

# for r in range(1, 6):
for r in [1, 3, 5, 10, 20]:
    print('Optmz rank', r)
    for e in ['test_5y', 'test_ally', 'train', 'ally']:
        d2 = df_optmz[(df_optmz['eval']==e) & (df_optmz['param']==f'normKGEr{r}') ]['kge'].values
        print(e, np.nanmean(d2), np.nanmedian(d2))
```

```python
# boxplot

train_met = []
for r in [1, 3, 5, 10, 20]:
    for e in [ 'train',]:
        d2 = df_optmz[(df_optmz['eval']==e) & (df_optmz['param']==f'normKGEr{r}') ]['kge'].values
        train_met.append(d2[ ~np.isnan(d2)])


test_met = []
for r in [1, 3, 5, 10, 20]:
    for e in [ 'test_5y',]:
        d2 = df_optmz[(df_optmz['eval']==e) & (df_optmz['param']==f'normKGEr{r}') ]['kge'].values
        test_met.append(d2[ ~np.isnan(d2)])

fig = plt.figure(figsize=[8, 4])

fig.add_subplot(1,2,1)
rank = [1, 3, 5, 10, 20]
plt.boxplot(train_met, showfliers=False)
plt.xticks(np.arange(len(rank))+1, rank)
plt.xlabel('Rank of calib parameter set')
plt.ylabel("KGE'")
plt.title('Train period')
plt.grid('on', linestyle='--')

fig.add_subplot(1,2,2)
rank = [1, 3, 5, 10, 20]
plt.boxplot(test_met, showfliers=False)
plt.xticks(np.arange(len(rank))+1, rank)
plt.xlabel('Rank of calib parameter set')
plt.ylabel("KGE'")
plt.title('Test period')
plt.grid('on', linestyle='--')

plt.tight_layout()
plt.show()
```

# How different are those parameters?

```python
tarbasin = 10
dfb = df_optmz[(df_optmz['basin']==tarbasin)]
kgetrain = []
kgetest = []
ranks = [1, 3, 5, 10, 20]
for r in ranks:
    kgetrain.append(dfb[ (dfb['param']==f'normKGEr{r}') & (dfb['eval']==f'train')]['kge'].values[0])
    kgetest.append(dfb[ (dfb['param']==f'normKGEr{r}') & (dfb['eval']==f'test_5y')]['kge'].values[0])

plt.plot(kgetrain, 'o-', label='train')
plt.plot(kgetest, 'o-', label='test')
plt.xticks(np.arange(len(ranks)), ranks)
plt.xlabel('parameter rank')
plt.ylabel("KGE'")
plt.legend()
plt.show()
```

```python
ranks = [1, 3, 5, 10, 20]
paramvalue = []
for r in ranks:
    files = glob.glob(f'/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/LongTermSimu/LSEallbasin/level1_{tarbasin}/normKGEr{r}/paramset_*.pkl')
    if len(files)>1:
        print('Warning!')
    else:
        file = files[0]
        dfr = pd.read_pickle(file)
        paramvalue.append( dfr['Value'].values)
paramvalue = np.array(paramvalue)
paramname = dfr['Parameter'].values


# load defa
file = f'/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/Calib_HH_MOASMO_bigrange/level1_{tarbasin}_MOASMOcalib/param_sets/paramset_iter0_trial0.pkl'
df = pd.read_pickle(file)
paramdefa = np.array([np.nanmean(i) for i in df['Value'].values])
paramvalue = paramvalue/paramdefa

# # normalize so they can be plotted in one figure
# paramvalue = ( paramvalue - np.nanmin(paramvalue,axis=0) ) / ( np.nanmax(paramvalue,axis=0) - np.nanmin(paramvalue,axis=0) )

fig = plt.figure(figsize=[12, 10])
for i in range(paramvalue.shape[1]):
    fig.add_subplot(4, 5, i+1)
    plt.plot(paramvalue[:,i], 'o-')
    plt.xticks(np.arange(len(ranks)), ranks)
    plt.xlabel('parameter rank')
    plt.title(paramname[i])
    
plt.tight_layout()
plt.show()
```
