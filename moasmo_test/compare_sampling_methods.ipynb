```python
# generate parameter scaling factors
num_init = 250
num_param = 15

factor_lh = sampling.lh(num_init, num_param)
factor_slh = sampling.slh(num_init, num_param)
factor_glp = sampling.glp(num_init, num_param)

factors = [factor_lh, factor_slh, factor_glp]
names = ['Latin Hypercube', 'Symmetric Latin Hypercube', 'Good Lattice Points']
```

```python
# plot figures

induse = [ [0, 1], [10, 11] ]
colors = ['k', 'b']

fig, axs = plt.subplots(2, 3, figsize=[10, 7])

for i in range(2):
    indi = induse[i]
    for j in range(3):
        axi = axs[i, j]
        axi.scatter(factors[j][:, indi[0]], factors[j][:, indi[1]], 10, color=colors[i])
        axi.set_title(names[j])
        axi.grid('on')
        axi.set_xlabel(f'factor {indi[0]}')
        axi.set_ylabel(f'factor {indi[1]}')

    
plt.tight_layout()
plt.show()
```
