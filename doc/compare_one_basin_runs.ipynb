```python
import xarray as xr
from matplotlib import pyplot as plt
import numpy as np
```

# Guoqiang VS Sean

```python
file1='/glade/scratch/guoqiang/CTSM_outputs/CAMELS_Calib/Lump_calib/archive/CAMELS_100/lnd/hist/CAMELS_100.clm2.h1.2000-01-01-00000.nc'
file2='/glade/scratch/swensosc/archive/hillslope_CAMELS_single/lnd/hist/hillslope_CAMELS_single.clm2.h1.2000-01-01-00000.nc'

ds1=xr.load_dataset(file1)
ds2=xr.load_dataset(file2)

ds1 = ds1.isel(lndgrid=100)
ds2 = ds2.isel(lndgrid=100)

ds1['time'] = ds1.indexes['time'].to_datetimeindex()
ds2['time'] = ds2.indexes['time'].to_datetimeindex()
```

```python
vall = []
for v in ds2.data_vars:
    if not v in ['date_written','time_written','time_bounds']:
        vall.append(v)
```

```python
fig, axs = plt.subplots(5, 4, figsize=[16, 16])
axs = axs.flatten()


for i in range(20):
    axi = axs[i]
    vi = vall[i]
    if ds1[vi].size>1:
        ds1[vi].plot(ax=axi, marker='o', markersize=3, linewidth=0.1, label='Guoqiang')
        ds2[vi].plot(ax=axi, linewidth=2, label='Sean')
    else:
        ds1[vi].plot(ax=axi, label='Guoqiang', alpha=0.5)
        ds2[vi].plot(ax=axi, label='Sean', alpha=0.5)
        
    axi.set_xlabel('')
    axi.set_title(vi)
    
    
    
    if i == 0:
        axi.legend()
    
plt.tight_layout()
plt.show()
```

```python
dd = ds1.QRUNOFF - ds2.QRUNOFF
dd.plot()
```
