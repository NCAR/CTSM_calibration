```python
import xarray as xr
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import os, sys, glob
```

```python

file_newHH = glob.glob(f'/glade/derecho/scratch/guoqiang/CTSM_outputs/CAMELS_Calib/Calib_HH_MOASMO/level1_0/level1_0_HHtest/run/level1_0_HHtest.clm2.h1.*.nc')
file_newHH_initinterp = glob.glob(f'/glade/derecho/scratch/guoqiang/CTSM_outputs/CAMELS_Calib/Calib_HH_MOASMO/level1_0/level1_0_HHtest_init_interp/run/level1_0_HHtest_init_interp.clm2.h1.*.nc')
file_oldHH = glob.glob(f'/glade/derecho/scratch/guoqiang/CTSM_outputs/CAMELS_Calib/Calib_HH_MOASMO/level1_0/level1_0_oldHHtest/run/level1_0_oldHHtest.clm2.h1.*.nc')

ds_newHH = xr.open_mfdataset(file_newHH)
ds_newHH_initinterp = xr.open_mfdataset(file_newHH_initinterp)
ds_oldHH = xr.open_mfdataset(file_oldHH)
```

```python
year = slice('2008', '2010')
d1 = ds_newHH.QRUNOFF.sel(time=year).values
d2 = ds_newHH_initinterp.QRUNOFF.sel(time=year).values
d3 = ds_oldHH.QRUNOFF.sel(time=year).values

fig = plt.figure(figsize=[10, 4])

fig.add_subplot(1,2,1)
plt.plot(d1, label='new')
plt.plot(d2, label='new init interp')
plt.legend()

fig.add_subplot(1,2,2)
plt.plot(d1, label='new')
plt.plot(d3, label='old')
plt.legend()

plt.tight_layout()
plt.show()
```
