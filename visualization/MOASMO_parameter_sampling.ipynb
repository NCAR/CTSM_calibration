```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import os, glob, sys
```

```python
# file = '/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/SA_HH_MOASMO/level1_1_SA/param_sets/paramset_iter0_scalefactors.csv'
# df = pd.read_csv(file)


params = np.nan * np.zeros([256, 12])

for i in range(256):
    file = f'/glade/campaign/cgd/tss/people/guoqiang/CTSM_CAMELS_proj/SA_HH_MOASMO/level1_1_SA/param_sets/paramset_iter0_trial{i}.pkl'
    df = pd.read_pickle(file)
    va = df['Value'].values
    for j in range(12):
        params[i, j]=np.nanmean(va[j])
```

```python
fig, axs = plt.subplots(12, 12, figsize=[12, 12])

for i in range(12):
    for j in range(12):
        ax = axs[i, j]
        
        if j>i:
            ax.axis('off')
            
        elif j==i:
            ax.hist(params[:,i], rwidth=0.7)
            
            # Hide X and Y axes label marks
            ax.xaxis.set_tick_params(labelbottom=False)
            ax.yaxis.set_tick_params(labelleft=False)

            # Hide X and Y axes tick marks
            ax.set_xticks([])
            ax.set_yticks([])
            
        else:
        
            ax.scatter(params[:,i], params[:,j], 0.1)

            # Hide X and Y axes label marks
            ax.xaxis.set_tick_params(labelbottom=False)
            ax.yaxis.set_tick_params(labelleft=False)

            # Hide X and Y axes tick marks
            ax.set_xticks([])
            ax.set_yticks([])
        
plt.tight_layout()
plt.show()
```
